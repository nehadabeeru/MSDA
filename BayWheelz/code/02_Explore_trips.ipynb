**San Francisco Trips**

```python
san_francisco.head()
```

```python
san_francisco.isnull().sum()
```

```python
# Plot the distribution of subscription type
sb.countplot(data = san_francisco, x = 'subscription_type', color = 'g')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.title('Distribution of Subscription Type in San Francisco')
```

```python
# Plot the distribution of start_station_name
plt.figure(figsize = [20, 10])
sb.countplot(data = san_francisco, y = 'start_station_name', color = 'r')
plt.xlabel('Count')
plt.ylabel('Start Station Name')
plt.title('Distribution of Start Station Name [san_francisco]')
```

```python
# Descrribe duration for san_francisco
san_francisco.duration.describe()
```

```python
san_francisco.duration /= 60
```

```python
# Descrribe duration for san_francisco
san_francisco.duration.describe()
```

Shortest Trip: 1 second

Longest Trip: 3536.516667 seconds (or 58.94 minutes)

```python
san_francisco['duration'].quantile(0.995)
```

99.5% of the duration values in the san_francisco DataFrame are less than or equal to approximately 86.20 seconds (~ 1.44 mins)

```python
# Convert to datetime
san_francisco.start_date = pd.to_datetime(san_francisco.start_date, format='%m/%d/%Y %H:%M')
```

```python
san_francisco.head()
```

```python
# Convert to datetime and adding to our trip
san_francisco['date'] = san_francisco.start_date.dt.date
```

```python
san_francisco.head()
```

```python
# Create a dict for each unique date in trip['date'] and the number of trips on that date
sf_trip_count = dict(san_francisco['date'].value_counts())
sf_trip_count
```

```python
# Create the dataframe using the sf_trip_count dict
sf_trip_count = pd.DataFrame.from_dict(sf_trip_count, orient='index').reset_index()
sf_trip_count.head()
```

```python
# Naming the columns
sf_trip_count.columns = ['date', 'trip_count']
```

```python
sf_trip_count.head(10)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['date']
  ys = series['trip_count']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = sf_trip_count.sort_values('date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('date')
_ = plt.ylabel('trip_count')
```

With this, we can figure out how many trips were taken on a particular date in SF

```python
# Sort the df_trip_count by date
sf_trip_count = sf_trip_count.sort_values(by='date')
```

```python
sf_trip_count.head()
```

```python
sf_trip_count.reset_index(drop = True, inplace = True)
```

```python
sf_trip_count.head(20)
```

```python
# Count the values in sf_trip_count
sf_trip_count.trip_count.value_counts()
```

```python
# Count the unique values of column date in train_main
sf_trip_count.date.nunique()
```

```python
sf_trip_count.info()
```

```python
plt.figure(figsize = (20, 15))
sf_trip_count.plot()
plt.title('Demand')
plt.ylabel('Trips')
plt.xlabel('Date')
plt.show()
```

```python
sf_trip_count.to_csv('../code/model/data/train_main_SF.csv', index=False)
```
